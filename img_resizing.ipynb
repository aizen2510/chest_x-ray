```python
import cv2 as cv
import os
import matplotlib.pyplot as plt

img_demo = r"C:\code\cnnt_model\chest_xray\test\PNEUMONIA\person99_bacteria_473.jpeg"


def imgResize():
    img = cv.imread(img_demo)
    if img is None:
        print("❌ Không thể đọc ảnh. Kiểm tra lại đường dẫn!")
        return

    # Chuyển ảnh từ BGR → RGB để hiển thị đúng màu trong matplotlib
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # ⚙️ Crop (cắt) vùng ảnh nếu muốn, cú pháp: img[y1:y2, x1:x2]
    img = img[100:400, 100:400]  # (Ví dụ crop lại 300x300 vùng trung tâm)

    height, width, _ = img.shape
    scale = 1 / 4  # thu nhỏ 4 lần

    # cac phuong phap noi suy
    interp_methods = [
        cv.INTER_AREA,
        cv.INTER_LINEAR,
        cv.INTER_NEAREST,
        cv.INTER_CUBIC,
        cv.INTER_LANCZOS4
    ]
    interp_titles = ['AREA', 'LINEAR', 'NEAREST', 'CUBIC', 'LANCZOS4']
    # Hiển thị ảnh gốc
    plt.figure(figsize=(10, 6))
    plt.subplot(2, 3, 1)
    plt.imshow(img)
    plt.title('Original')
    plt.axis('off')
    # Resize với các phương pháp khác nhau
    for i, method in enumerate(interp_methods):
        img_resize = cv.resize(img, (int(width * scale), int(height * scale)), interpolation=method)
        plt.subplot(2, 3, i + 2)
        plt.imshow(img_resize)
        plt.title(interp_titles[i])
        plt.axis('off')

    plt.tight_layout()
    plt.show()


# Gọi hàm
imgResize()
```
